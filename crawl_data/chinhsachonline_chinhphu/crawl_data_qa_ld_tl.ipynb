### I. Import library

```python
import pandas as pd
from crawl_utils import crawl_page
```

### II. Crawl data 

#### 2.1. Excute

```python
articles_data = []

for page_num in range(1, 600): 
    print(page_num)
    page_url = f"https://chinhsachonline.chinhphu.vn/lao-dong-tien-luong/t1/11/trang-{page_num}.htm"
    articles_data.extend(crawl_page(page_url))
```

#### 2.2. Export CSV

```python
df = pd.DataFrame(articles_data)
df.head()
```

```python
csv_file = './data/laodong_tienluong.csv'
df.to_csv(csv_file, encoding='utf-8',index_label='Index')
```
